```python
import os
import ujson
import pandas as pd
from pathlib import Path
from tqdm import tqdm
```

# Config

```python
result_path = Path(os.getcwd()).parent / "results"
output_path = Path(os.getcwd()).parent / "analysis_results"
```

# Result Compilation

```python
result_summary = {}
summary_file_name = "summary.json"
```

```python
for llm in tqdm(os.listdir(result_path)):
    if os.path.isdir(result_path / llm):
        result_summary[llm] = {}
        for task in tqdm(os.listdir(result_path / llm)):
            if os.path.isdir(result_path / llm / task):
                result_summary[llm][task] = {}
                for pe in tqdm(os.listdir(result_path / llm / task)):
                    if os.path.isdir(result_path / llm / task / pe):
                        result_summary[llm][task][pe] = {}
                        with open(result_path / llm / task / pe / summary_file_name, "r") as f:
                            result_summary[llm][task][pe] = ujson.load(f)
```

```python
result_summary
```

```python
for model in tqdm(result_summary.keys()):
    for task in tqdm(result_summary[model].keys()):
        for pe in tqdm(result_summary[model][task].keys()):
            lengths = []
            time_takens = []
            files = [file for file in os.listdir(result_path / model / task / pe) if
                     file != summary_file_name and file != ".DS_Store"]
            for file in tqdm(files):
                with open(Path(result_path) / model / task / pe / file) as f:
                    data = ujson.load(f)
                    lengths.append(data["length"])
                    time_takens.append(data["time_taken"])
            mean_length = sum(lengths) / len(lengths)
            mean_time_taken = sum(time_takens) / len(time_takens)
            std_length = (sum([(length - mean_length) ** 2 for length in lengths]) / len(lengths)) ** 0.5
            std_time_taken = (sum([(time_taken - mean_time_taken) ** 2 for time_taken in time_takens]) / len(
                time_takens)) ** 0.5
            min_length = min(lengths)
            min_time_taken = min(time_takens)
            max_length = max(lengths)
            max_time_taken = max(time_takens)
            median_length = sorted(lengths)[len(lengths) // 2]
            median_time_taken = sorted(time_takens)[len(time_takens) // 2]
            total_time_taken = sum(time_takens)
            result_summary[model][task][pe]["mean_length"] = mean_length
            result_summary[model][task][pe]["median_length"] = median_length
            result_summary[model][task][pe]["std_length"] = std_length
            result_summary[model][task][pe]["min_length"] = min_length
            result_summary[model][task][pe]["max_length"] = max_length
            result_summary[model][task][pe]["mean_time_taken"] = mean_time_taken
            result_summary[model][task][pe]["median_time_taken"] = median_time_taken
            result_summary[model][task][pe]["std_time_taken"] = std_time_taken
            result_summary[model][task][pe]["min_time_taken"] = min_time_taken
            result_summary[model][task][pe]["max_time_taken"] = max_time_taken
            result_summary[model][task][pe]["total_time_taken"] = total_time_taken
```

```python
df = pd.DataFrame.from_dict({(i, j, k): result_summary[i][j][k]
                             for i in result_summary.keys()
                             for j in result_summary[i].keys()
                             for k in result_summary[i][j].keys()},
                            orient='index')
```

```python
df = df.drop(columns=["current_index", "is_processing"])
```

## Null-Shot vs Zero-Shot

```python
null_shot_zero_shot_df = df[((df['pe_technique'] == 'zero-shot') | (df['pe_technique'] == 'null-shot')) & ~(
        df['model'].str.startswith('pythia') | df['model'].str.startswith('qwen'))]
null_shot_zero_shot_df
```

```python
def print_relative_result(data_df, tasks_order,
                          models_order=None, pe_compare=('zero-shot', 'null-shot')):
    if models_order is None:
        models_order = ['palm-2-text', 'palm-2-chat', 'gemini-pro-text', 'gemini-pro-chat',
                        'gpt-3.5-turbo', 'gpt-4-turbo', 'claude-2.1', 'claude-3-haiku',
                        'claude-3-sonnet', 'claude-3-opus']
    accuracy_change = {}
    for m in models_order:
        accuracy_change[m] = {}
        for t in tasks_order:
            zero_shot_accuracy = data_df[
                (data_df['model'] == m) & (data_df['task'] == t) & (
                        data_df['pe_technique'] == pe_compare[0])]['accuracy'].values[0]
            null_shot_accuracy = data_df[
                (data_df['model'] == m) & (data_df['task'] == t) & (
                        data_df['pe_technique'] == pe_compare[1])]['accuracy'].values[0]
            accuracy_change[m][t] = {
                pe_compare[0]: round(zero_shot_accuracy, 2),
                pe_compare[1]: round(null_shot_accuracy, 2),
                'percentage_change': round((null_shot_accuracy - zero_shot_accuracy) / zero_shot_accuracy * 100, 2)
            }
    accuracy_change = dict(sorted(accuracy_change.items(),
                                  key=lambda x: models_order.index(x[0])))
    for m in accuracy_change.keys():
        accuracy_change[m] = dict(sorted(accuracy_change[m].items(),
                                         key=lambda x: tasks_order.index(x[0])))

    model_map = {
        'palm-2-text': 'PaLM 2',
        'palm-2-chat': 'PaLM 2 (Chat)',
        'gemini-pro-text': 'Gemini 1.0 Pro',
        'gemini-pro-chat': 'Gemini 1.0 Pro (Chat)',
        'gpt-3.5-turbo': 'GPT-3.5 Turbo',
        'gpt-4-turbo': 'GPT-4 Turbo',
        'claude-2.1': 'Claude 2.1',
        'claude-3-haiku': 'Claude 3 Haiku',
        'claude-3-sonnet': 'Claude 3 Sonnet',
        'claude-3-opus': 'Claude 3 Opus',
    }

    string_to_print = ""
    for m in accuracy_change.keys():
        string_to_print += model_map[m] + " & "
        for t in accuracy_change[m].keys():
            if accuracy_change[m][t]['percentage_change'] > 0:
                string_to_print += f"\\textcolor{{mygreen}}{{{accuracy_change[m][t]['percentage_change']}\\%}} & "
            else:
                string_to_print = string_to_print + f"{accuracy_change[m][t]['percentage_change']}\\% & "
        string_to_print = string_to_print[:-3] + "\\\\\n"
    print(string_to_print)
```

```python
def print_absolute_result(data_df, tasks_order, models_order=None, pe_compare=('zero-shot', 'null-shot')):
    if models_order is None:
        models_order = ['palm-2-text', 'palm-2-chat', 'gemini-pro-text', 'gemini-pro-chat',
                        'gpt-3.5-turbo', 'gpt-4-turbo', 'claude-2.1', 'claude-3-haiku',
                        'claude-3-sonnet', 'claude-3-opus']
        
    accuracy_change = {}
    for model in models_order:
        accuracy_change[model] = {}
        for task in tasks_order:
            zero_shot_accuracy = data_df[
                (data_df['model'] == model) & (data_df['task'] == task) & (
                        data_df['pe_technique'] == pe_compare[0])]['accuracy'].values[0]
            null_shot_accuracy = data_df[
                (data_df['model'] == model) & (data_df['task'] == task) & (
                        data_df['pe_technique'] == pe_compare[1])]['accuracy'].values[0]
            accuracy_change[model][task] = {
                pe_compare[0]: round(zero_shot_accuracy, 2),
                pe_compare[1]: round(null_shot_accuracy, 2),
                'percentage_change': round((null_shot_accuracy - zero_shot_accuracy) / zero_shot_accuracy * 100, 2)
            }
    accuracy_change = dict(sorted(accuracy_change.items(),
                                  key=lambda x: models_order.index(x[0])))
    for m in accuracy_change.keys():
        accuracy_change[m] = dict(sorted(accuracy_change[m].items(),
                                         key=lambda x: tasks_order.index(x[0])))
        
    model_map = {
        'palm-2-text': 'PaLM 2',
        'palm-2-chat': 'PaLM 2 (Chat)',
        'gemini-pro-text': 'Gemini 1.0 Pro',
        'gemini-pro-chat': 'Gemini 1.0 Pro (Chat)',
        'gpt-3.5-turbo': 'GPT-3.5 Turbo',
        'gpt-4-turbo': 'GPT-4 Turbo',
        'claude-2.1': 'Claude 2.1',
        'claude-3-haiku': 'Claude 3 Haiku',
        'claude-3-sonnet': 'Claude 3 Sonnet',
        'claude-3-opus': 'Claude 3 Opus',
    }

    string_to_print = ""
    for m in accuracy_change.keys():
        string_to_print += model_map[m] + " & "
        for t in accuracy_change[m].keys():
            if accuracy_change[m][t]['percentage_change'] > 0:
                string_to_print += f"{accuracy_change[m][t][pe_compare[0]]} & \\textcolor{{mygreen}}{{{accuracy_change[m][t][pe_compare[1]]}}} & "
            else:
                string_to_print = string_to_print + f"{accuracy_change[m][t][pe_compare[0]]} & {accuracy_change[m][t][pe_compare[1]]} & "
        string_to_print = string_to_print[:-3] + "\\\\\n"
    print(string_to_print)
```
